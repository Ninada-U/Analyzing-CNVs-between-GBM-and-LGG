```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10, 8
```

```python
lgg_p = 'LGG.focal_score_by_genes.txt'
gbm_p = 'GBM.focal_score_by_genes.txt'
```

```python
lgg_df = pd.read_csv(lgg_p, sep = '\t')
gbm_df = pd.read_csv(gbm_p, sep = '\t')
```

```python
# lgg_df.head(2)
```

```python
# gbm_df.head(2)
```

```python
plt.imshow(lgg_df.iloc[:, 3:].T, aspect='auto')
plt.title('LGG focal score by gene')
plt.xlabel('Gene Indices')
plt.ylabel('Case Indices')
plt.colorbar()
plt.savefig('LGG focal.png')
```

```python
plt.imshow(gbm_df.iloc[:, 3:].T, aspect='auto')
plt.title('GBM Focal Score by Gene')
plt.xlabel('Gene Indices')
plt.ylabel('Case Indices')
plt.colorbar()
plt.savefig('GBM focal.png')
```

```python
x = gbm_df.iloc[:, 3:].T.columns
pos = gbm_df.iloc[:, 3:].T.apply(lambda c: len(c[c>0]))
neg = gbm_df.iloc[:, 3:].T.apply(lambda c: len(c[c<0]))
plt.scatter(x, pos, s=.2, color='blue')
plt.scatter(x, neg, s=.2, color='red')
plt.title('GBM Manhattan Plot')
plt.xlabel('Gene Indices')
plt.ylabel("Sum of Focal Scores")
```

```python
x = gbm_df.iloc[:, 3:].T.columns
pos = gbm_df.iloc[:, 3:].T.apply(lambda c: len(c[c>0]))
neg = gbm_df.iloc[:, 3:].T.apply(lambda c: len(c[c<0])) * -1
plt.scatter(x, pos, s=.2, color='blue')
plt.scatter(x, neg, s=.2, color='red')
plt.title('GBM Focal Score Sums by Gene Indices')
plt.xlabel('Gene Indices')
plt.ylabel("Focal Score Counts")
```
